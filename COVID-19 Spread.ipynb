```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
class contagiado:             
    def __init__(self,t_c):    
        self.t_c = t_c             
        self.N_c = 0     
    def is_new_cont(self,t,time_for_new_cont):             
        if (t-self.t_c)/time_for_new_cont > (self.N_c+1) and (t-self.t_c) <14:
            self.N_c += 1
            return True     
```

```python
R0 = 3.0 # Virus transmitability scale
days_sim = 30   # total days to run simulation
time_for_new_cont = 14./R0
```

```python
Joe = contagiado(0)
arr_cont = np.array([Joe])

```

```python
# Run Simulation now
t_c = np.array([0])
N_cum = np.array([1])
N_c = 1
for t in np.arange(0,days_sim,time_for_new_cont/40.):
    for cont in arr_cont:
        is_new_cont = cont.is_new_cont(t,time_for_new_cont)                                                                               
        if is_new_cont:
            N_c += 1
            new_cont = contagiado(t)
            arr_cont = np.append(arr_cont,new_cont)
            t_c = np.append(t_c,t)
            N_cum = np.append(N_cum,N_c)
    print("t = %.2f days" % t)
    print("N_infected =",N_c)            
```

```python
plt.figure(figsize=(10,7))
plt.plot(t_c,N_cum,'-k')
plt.yscale('log')
plt.xlabel('days since first infected',size='large')
plt.ylabel('Number of infected',size='large')
```
